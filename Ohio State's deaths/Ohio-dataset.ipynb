```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
cause = "Chronic Disease","Alzheimer","TB","Accident","Hepatitas","Cardiac Arrest"
percentile = [13,6,1,60,15,5]
```

```python
plt.figure(figsize=(5,5))
space = (0.02,0.02,0.02,0.05,0.02,0.02)
```

```python
plt.pie(percentile,labels=cause,explode=space,startangle=70)
plt.axis("equal")
plt.title("Ohio State - Leading death causesvfrom 2012")
plt.show()
```
